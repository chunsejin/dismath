```python
# 답안 code
# 주석 작성 필요
def Hamilton_graph(g, source):
  li = []
  visited_Node = []
  g_Edges = list(g.edges)
  for edge in g_Edges: # List화 
      temp = list(edge)
      temp.append(False)
      li.append(temp)
  edges_list = get_Edge_List(source, li)
  backTracking(g, edges_list, source, source,-1, li, visited_Node)

def get_Edge_List(source, li):
  edges_list = [[val, truth] for (node, val, truth) in li if node is source]
  and_list = [[node, truth] for (node, val, truth) in li if val is source]
  edges_list.extend(and_list)
  return edges_list

def backTracking(g, edges_list, source, first_Node, lately_node, li, visited_Node):
  if len(visited_Node) >= len(g.nodes) and first_Node == source:
    # visited_Node.insert(0, source)
    # print(visited_Node)
    print("Hamilton Graph")
    return True

  for node,truth in edges_list:
    if truth == False and node != lately_node:
      visited_Node.append(node)
      for i in range(len(li)):
        if li[i][0] == source and li[i][1] == node:
          li[i][2] = True
        if li[i][0] == node and li[i][1] == source:
          li[i][2] = True
      edge_list = get_Edge_List(node,li)
      flag = backTracking(g, edge_list, node, first_Node, source, li, visited_Node)
      if flag is True:
        return True

      for i in range(len(li)):
        if li[i][0] == source and li[i][1] == node:
          li[i][2] = False
        if li[i][0] == node and li[i][1] == source:
          li[i][2] = False
      visited_Node.pop(len(visited_Node)-1)
  if len(visited_Node) == 0 :
    print("This is not Hamilton Graph")
    return False
```
